```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from astropy.io import fits
from astropy.visualization import make_lupton_rgb
import matplotlib.pyplot as plt
```

```python
data_blue = fits.getdata('data/NGC5053_blue.fits')
data_red = fits.getdata('data/NGC5053_red.fits')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,7))
ax1.imshow(data_blue, vmin=0, vmax=50, cmap='Greys_r', origin='lower')
ax2.imshow(data_red, vmin=0, vmax=50, cmap='Greys_r', origin='lower')
```

```python
data_green = (data_red + data_blue)/2
```

```python
fig, ax = plt.subplots(1, figsize=(7,7))
ax.imshow(data_green, vmin=0, vmax=50, cmap='Greys_r', origin='lower')
```

```python
color_image = make_lupton_rgb(data_red, data_green, data_blue,
                              minimum=0, stretch=10)

fig, ax = plt.subplots(1, figsize=(15,15))
ax.imshow(color_image, origin='lower')
```
